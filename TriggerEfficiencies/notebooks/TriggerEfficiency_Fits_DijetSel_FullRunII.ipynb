```python
import mplhep as hep
import matplotlib.pyplot as plt
hep.style.use("CMS")
%matplotlib inline

def do_fit(x,y,):
    #linear fit of fitted trigger turn-on values vs. nominal pT value of trigger
    coef = np.polyfit(x,y,1)
    poly1d_fn = np.poly1d(coef) 
    return poly1d_fn
```

```python
year_list = ['2016_preVFP','2016','2017','2018']

for year in year_list:

    outputDir = f'../Plots_Aug23_dijetSel/TriggerStudies/{year}/'
    if not(os.path.exists(outputDir)): 
        os.makedirs(outputDir)


    CMS_lumi.lumi_13TeV = f"{year} (13 TeV)"
    #units = [/micro barns] 
    lumiTot = triggerTable['AK8PFJet500']['luminosity'][year]
    lumiRecorded = []
    triggerDenomList = ['AK8PFJet500','AK8PFJet450','AK8PFJet400','AK8PFJet320','AK8PFJet260','AK8PFJet200','AK8PFJet140','AK8PFJet80','AK8PFJet60']
    triggerPassList = ['AK8PFJet550','AK8PFJet500','AK8PFJet450','AK8PFJet400','AK8PFJet320','AK8PFJet260','AK8PFJet200','AK8PFJet140','AK8PFJet80'] 
    nameList = ['AK8PFJet550','AK8PFJet500','AK8PFJet450','AK8PFJet400','AK8PFJet320','AK8PFJet260','AK8PFJet200','AK8PFJet140','AK8PFJet80'] 



    if '2016' in year: 
        triggerDenomList = triggerDenomList[1:]
        triggerPassList = triggerPassList[1:]
        nameList = nameList[1:]

    for it in triggerPassList:
        lumiRecorded.append(triggerTable[it]['luminosity'][year])
    
    
    prescale=list(lumiTot/np.array(lumiRecorded))
    print(year,lumiRecorded)
    print(triggerPassList)
    print(prescale)

    lumi=lumiTot/(10**9)

    triggerDictAllYears = OrderedDict()
    triggerDictAllYears[year] = OrderedDict()

    triggerDict = OrderedDict()
    for it,ipresc in zip(nameList,prescale):
        triggerDict[it]=dict()
        triggerDict[it]['threshold'] = it.split('PFJet')[1]
        triggerDict[it]['prescale'] = ipresc
        triggerDict[it]['luminosity'] = lumi/ipresc
        triggerDict[it]['turn-on_1'] = 0.
        triggerDict[it]['turn-on_2'] = 0.


    inFileSample=ROOT.TFile.Open(f'../TrigEff_Plots/triggerEfficiencies_histograms_MiniAOD_JetHTRun{year}ALL_eta2pt4.root','r')
    sample=f'JetHT{year}'
    
    
    ########################################################################## 
    ######## Carry out efficiency curve fits to get trigger turn-ons  ########
    ##########################################################################
    
    
    # make a variety of plots for validations
    triggerDict_tmp = fitTriggerEfficiencies_perYear(inFileSample, sample, year, triggerDict, triggerPassList, triggerDenomList, nameList ,lumi=lumi, outputDir=outputDir,
                                                               minFit=60., maxFit=1200., ext='.png', version='_Aug23_Unzoomed', xlabel='Leading jet pt [GeV]', 
                                                               xmin=50, xmax=1500, rebin=1, fitRefine=False, unzoomed=True)
    
    triggerDict_tmp = fitTriggerEfficiencies_perYear(inFileSample, sample, year, triggerDict, triggerPassList, triggerDenomList, nameList ,lumi=lumi, outputDir=outputDir,
                                                               minFit=60., maxFit=1200., ext='.png', version='_Aug23_FinerFit_Unzoomed', xlabel='Leading jet pt [GeV]', 
                                                               xmin=50, xmax=1500, rebin=1, fitRefine=True, unzoomed=True)
    
    triggerDict_tmp = fitTriggerEfficiencies_perYear(inFileSample, sample, year, triggerDict, triggerPassList, triggerDenomList, nameList ,lumi=lumi, outputDir=outputDir,
                                                               minFit=60., maxFit=1200., ext='.pdf', version='_Aug23_Unzoomed', xlabel='Leading jet pt [GeV]', 
                                                               xmin=50, xmax=1500, rebin=1, fitRefine=False, unzoomed=True)
    
    triggerDict_tmp = fitTriggerEfficiencies_perYear(inFileSample, sample, year, triggerDict, triggerPassList, triggerDenomList, nameList ,lumi=lumi, outputDir=outputDir,
                                                               minFit=60., maxFit=1200., ext='.pdf', version='_Aug23_FinerFit_Unzoomed', xlabel='Leading jet pt [GeV]', 
                                                               xmin=50, xmax=1500, rebin=1, fitRefine=True, unzoomed=True)
    
    
    
    triggerDict_tmp = fitTriggerEfficiencies_perYear(inFileSample, sample, year, triggerDict, triggerPassList, triggerDenomList, nameList ,lumi=lumi, outputDir=outputDir,
                                                               minFit=60., maxFit=1200., ext='.png', version='_Aug23', xlabel='Leading jet pt [GeV]', 
                                                               xmin=50, xmax=1500, rebin=1, fitRefine=False)
    
    triggerDict_tmp = fitTriggerEfficiencies_perYear(inFileSample, sample, year, triggerDict, triggerPassList, triggerDenomList, nameList ,lumi=lumi, outputDir=outputDir,
                                                               minFit=60., maxFit=1200., ext='.pdf', version='_Aug23', xlabel='Leading jet pt [GeV]', 
                                                               xmin=50, xmax=1500, rebin=1, fitRefine=False)
    
    
    triggerDict_tmp = fitTriggerEfficiencies_perYear(inFileSample, sample, year, triggerDict, triggerPassList, triggerDenomList, nameList ,lumi=lumi, outputDir=outputDir,
                                                               minFit=60., maxFit=1200., ext='.pdf', version='_Aug23', xlabel='Leading jet pt [GeV]', 
                                                               xmin=50, xmax=1500, rebin=1)
    del (triggerDict_tmp)
    gc.collect()
    # main plots/fits, use fitted turnons from the below line in the following linear fit of turn on values
    triggerDict = fitTriggerEfficiencies_perYear(inFileSample, sample, year, triggerDict, triggerPassList, triggerDenomList, nameList ,lumi=lumi, outputDir=outputDir,
                                                               minFit=60., maxFit=1200., ext='.png', version='_Aug23_FinerFit', xlabel='Leading jet pt [GeV]', 
                                                               xmin=50, xmax=1500, rebin=1, verbose=True)
    
    triggerDictAllYears[year] = copy.deepcopy(triggerDict)
    
    ########################################################################## 
    ####### Carry out linear fit of turn-ons obtained from fits  above #######
    ##########################################################################
    
    fig, ax = plt.subplots(figsize=(7.5,5),dpi=200)#300)

    turnon_1 = []
    turnon_2 = []
    thresholds = []
    for it in reversed(triggerDict.keys()):
        turnon_1.append(triggerDict[it]['turn-on_1'])
        turnon_2.append(triggerDict[it]['turn-on_2'])
        thresholds.append(int(triggerDict[it]['threshold']))
        
    plt.plot(thresholds[:],turnon_2[:],'ko',label=r'$f_{eff,2}$ : $p_T^{reco}|_{f=0.95N} <p_T^{reco} < 3.5*x$',markersize=5,markerfacecolor='none')
    plt.plot(thresholds[:],turnon_1[:],'bX',label=r'$f_{eff,1}$ : x/1.1 < $p_T^{reco}$ < 3.5*x',markersize=5,markerfacecolor='none')

    y1, r1, *_ = np.polyfit(thresholds[:],turnon_1[:],1,full=True)
    y2, r2, *_ = np.polyfit(thresholds[:],turnon_2[:],1,full=True)


    fit1 = np.poly1d(y1)
    fit2 = np.poly1d(y2)

    plt.plot(thresholds[:],fit2(thresholds[:]),'r--',label=r'$f_{eff,2}$'+f" fit: $p_0={y2[0]:0.3f},p_1={y2[1]:0.3f},(\chi^2/ndf$={r2[0]:0.2f}/{len(thresholds[:])-1-1})")
    plt.plot(thresholds[:],fit1(thresholds[:]),'--',color='maroon',label=r'$f_{eff,1}$'+f" fit: $p_0={y1[0]:0.3f},p_1={y1[1]:0.3f},(\chi^2/ndf$={r1[0]:0.2f}/{len(thresholds[:])-1-1})")

    for x,y in zip(thresholds[:],fit1(thresholds[:])):
        plt.text(x-11,y-22,f'{str(x)}, {str(np.round(y,2))}', color='maroon', fontsize=5)

    for x,y in zip(thresholds[:],fit2(thresholds[:])):
        plt.text(x-27,y+12,f'{str(x)}, {str(np.round(y,2))}', color='red', fontsize=5)

    print(fit1,fit2)
    print(fit1(thresholds[:]),fit2(thresholds[:]))
    hep.cms.lumitext(r"AK8 PUPPI, %.2f fb$^{-1}$"%(lumiTot/(10**9))+f" ({year}), 13 TeV",fontsize=14)
    hep.cms.text(r"Preliminary",fontsize=14,loc=2)
    ax.legend(fontsize=9,loc=(0.44,0.05))
    ax.set_xlabel(r'Trigger threshold [GeV]', loc='right',fontsize=14)
    ax.set_ylabel(r'99.5% efficiency $p_T$ [GeV]', loc='top',fontsize=14)
    ax.tick_params(axis='both', which='major', labelsize=14)
    ax.tick_params(axis='both', which='minor', labelsize=14)#plt.yscale('log')

    plt.ylim(60.,800. if not '2016' in year else 700.)
    plt.xlim(30.,600. if not '2016' in year else 550.)
    plt.grid(200)
    plt.savefig(outputDir+f'TriggerThresholds_vs_99pc_pT_UL{year}.png')
    plt.savefig(outputDir+f'TriggerThresholds_vs_99pc_pT_UL{year}.pdf')
    plt.show()
    del(triggerDict)
    inFileSample.Close()
    gc.collect()
```
