### in 2017/2018 as per the btagSFProducer in nanoAOD tools: wp_btv = { "l" : 0, "m" : 1, "t" : 2, "shape_corr" : 3 }; where the shape corr is not present in these 2016 files. Further, flavours for AK4 jets in the 2016 csvs are what one would expect them to be labeled as, but the old scheme (as seen above in 2017/2018) relabeled hadronFlavours as folows: 5=>0,4=>1,[ 0(, 1, 2, 3, 21) ]=>2
#### so, relabeling L as 0, M as 1, T as 2 in 2016 CSVs;
#### and relabeling the hadronFlavours for 2016,  as done for 17/18 a la old btv CSV UL format used in btagSFProducer https://github.com/cms-nanoAOD/nanoAOD-tools/blob/acf3bf574e7fb9a976159ceee67c2ca079907fdb/python/postprocessing/modules/btv/btagSFProducer.py#L308

### DeepJet .csv WP file label modification for 2016_preVFP

```python
fn_csv_2016_preVFP = 'DeepJet_106XUL16preVFPSF_v1.csv'
file_2016_preVFP = pd.read_csv(fn_csv_2016_preVFP)
```

```python
file_2016_preVFP
```

```python
file_2016_preVFP.keys()
```

```python
print(file_2016_preVFP['jetFlavor'][0:50])
```

```python
file_2016_preVFP.columns=file_2018.columns
for i,WP in enumerate(file_2016_preVFP['DeepJet;OperatingPoint']):
    
    if "L" in WP: 
        file_2016_preVFP.loc[i,'DeepJet;OperatingPoint']=0
    elif "M" in WP:
        file_2016_preVFP.loc[i,'DeepJet;OperatingPoint']=1
    elif "T" in WP:
        file_2016_preVFP.loc[i,'DeepJet;OperatingPoint']=2



for i,flav in enumerate(file_2016_preVFP[' jetFlavor']):
    
    if (abs(flav)==5): #b
        file_2016_preVFP.loc[i,' jetFlavor']=0
    elif (abs(flav)==4): #c
        file_2016_preVFP.loc[i,' jetFlavor']=1
    elif (abs(flav) in [0,1,2,3,21]): #udsg
        file_2016_preVFP.loc[i,' jetFlavor']=2
        
#file_2016_preVFP.columns=file_2018.columns        
```

```python
file_2016_preVFP
```

```python
file_2016_preVFP.to_csv('DeepJet_106XUL16preVFPSF_v1_modified.csv', index=False)
```
